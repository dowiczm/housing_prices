```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def re_name(df):
    news = []
    for i in df.columns:
        col = i.lower()
        splits = (col.split())
        joint = '_'.join(splits)
        news.append(joint)
    return news
```

```python
def null_check(df, cols):
    nulls = []
    for col in cols:
        if df[col].isnull().sum() > 0:
            nulls.append((col, df[col].isnull().sum()))
    return nulls
```

```python
def fix_nulls(df):
    nulls = []
    for i in df.columns:
        if df[i].isnull().sum() > 0:
            nulls.append(i)
    num_columns = []
    object_columns = []
    for i in nulls:
        if df[i].dtype != 'object':
            num_columns.append(i)
        else:
            object_columns.append(i)
    for i in num_columns:
        df[i].fillna(0, inplace=True)
    for i in object_columns:
        df[i].fillna('None', inplace=True)
    return  
```

```python
test = pd.read_csv('datasets/test.csv')
test.head()
```

```python
#matching columns names to correspond with base_train
new_columns = re_name(test)
test.columns = new_columns
test.head()
```

```python
null_check(test, test.columns)
```

```python
#using same function used on base_train to impute nulls 
fix_nulls(test)
null_check(test, test.columns)
```
